# 聚类-高斯混合模型 

## 前提知识

### 条件概率

条件概率既是指在事件空间 $\Omega$ 中，当某个事件 $A$ 发生的前提下，且$P(A) > 0$，另一个事件 $B$ 发生的概率。
$$
P(B \mid A) = \frac{P(AB)}{P(A)}
$$

通常在没有约束条件的情况下，事件 $B$ 的概率 $P(B) = \frac{|B|}{|\Omega|}$。但是如果考虑事件 $A$ 发生的前提下，则将事件 $B$ 所参照的事件空间 $\Omega$ 变为 $A$，且 $A$ 是 $\Omega$ 的非空子集，$B$ 与 $A$ 的交集 $A \cap B$，这时 $B$ 相对于前提条件 $A$ 的概率为：
$$
P(B \mid A) = \frac{|AB|}{|A|} = \frac{\frac{|AB|}{|\Omega|}}{\frac{|A|}{|\Omega|}} = \frac{P(AB)}{P(A)}
$$


也可以从乘法公式出发来理解，想要知道事件 $A,B$ 同时发生的概率，可以从两个方向考虑：

1. 第一步确定 $A$ 发生的概率，第二步确定在事件 $A$ 发生的前提下，事件 $B$ 发生的概率；
   
2. 第一步确定 $B$ 发生的概率，第二步确定在事件 $B$ 发生的前提下，事件 $A$ 发生的概率。
   
这两个方向是等价，即
$$P(AB)=P(A)P(B \mid A)=P(B)P(A \mid B)$$

### 全概率公式

若事件空间 $\Omega$ 可以被分割为 $B_1, B_2,\cdots,B_n$ 个互不相容的事件，且有 $B_1 \cup B_2 \cup \cdots \cup B_n=\Omega$ ，如果 $P(B_i)>0，i=1,2,\cdots,n$，则对任一事件 $A$ 有
$$
P(A)=\sum_{i=1}^n P(B_i)P(A \mid B_i)
$$

全概率公式表明：求事件 $A$ 的发生受到多种因素 $B_i$ 影响，那么事件 $A$ 发生的概率则是每种因素影响下事件 $A$ 发生概率的总和。

### 贝叶斯公式
通过全概率公式可以通过综合考虑各种因素的影响计算某个事件 $A$ 发生的概率。现在反过来，如果事件 $A$ 发生了，究竟是哪种因素影响最大呢？贝叶斯公式可以处理这个问题。

若事件空间 $\Omega$ 可以被分割为 $B_1, B_2,\cdots,B_n$ 个互不相容的事件，且有 $B_1 \cup B_2 \cup \cdots \cup B_n=\Omega$ ，如果$P(A) > 0$， $P(B_i)>0，i=1,2,\cdots,n$，则有

$$
P(B_i \mid A)=\frac{P(B_i)P(A \mid B_i)}{P(A)}
$$

- $P(A \mid B_i)$ 是已知 $B_i$ 发生后 $A$ 的条件概率，也被称作 $A$ 的后验概率。有时称为 $B_i$ 的似然性。

- $P(B_i \mid A)$ 是已知 $A$ 发生后 $B_i$ 的条件概率，也称作 $B_i$ 的后验概率。

- $P(A)$ 是 $A$ 的先验概率。之所以称为"先验"是因为它不考虑任何 $B_i$ 方面的因素。

- $P(B_i)$ 是 $B_i$ 的先验概率，也作标准化常量（normalized constant）。

贝叶斯定理可以简单理解为：

**后验概率 = (似然性*先验概率)/标准化常量**

> 基于贝叶斯定理：即使100%的胰腺癌症患者都有某症状，而某人有同样的症状，绝对不代表该人有100%的概率得胰腺癌，还需要考虑先验概率，假设胰腺癌的发病率是十万分之一，而全球有同样症状的人有万分之一，则此人得胰腺癌的概率只有十分之一，90%的可能是是假阳性。

### 大数定律（Law of Large Numbers, LLN）

如果总体的均值存在，则随着样本数量的增加，样本均值依概率收敛于总体均值。换言之，事件发生的频率依概率收敛于事件的总体概率。

$$
 {\overline {X}}_{n}\ {\xrightarrow {P}}\ \mu \quad {\text {当}}\quad n\to \infty 
$$
即
$$
 \lim_{n \to \infty } P(|{\overline {X}}_{n}-\mu |>\varepsilon)=0
$$

大数定律很重要，因为它“说明”了一些随机事件的均值的长期稳定性。人们发现，在重复试验中，随着试验次数的增加，事件发生的频率趋于一个稳定值；人们同时也发现，在对物理量的测量实践中，测定值的算术平均也具有稳定性。比如，我们向上抛一枚硬币，硬币落下后哪一面朝上是偶然的，但当我们上抛硬币的次数足够多后，达到上万次甚至几十万几百万次以后，我们就会发现，硬币每一面向上的次数约占总次数的二分之一，亦即偶然之中包含着必然。

投硬币，正面1， 反面为0， 所有结果的均值趋于0.5。

### 中心极限定理（Central Limit Theorem， CLT）
中心极限定理是概率论中的一个重要定理。中心极限定理说明，在适当的条件下，大量相互独立随机变量的均值经适当标准化后依分布收敛于标准正态分布。这组定理是数理统计学和误差分析的理论基础，指出了大量随机变量之和近似服从正态分布的条件。

> 中心极限定理有着有趣的历史。这个定理的第一版被法国数学家棣莫弗发现，他在1733年发表的卓越论文中使用正态分布去估计大量抛掷硬币出现正面次数的分布。这个超越时代的成果险些被历史遗忘，所幸著名法国数学家拉普拉斯在1812年发表的巨著 Théorie Analytique des Probabilités中拯救了这个默默无名的理论。

> 拉普拉斯扩展了棣莫弗的理论，指出二项分布可用正态分布逼近。但同棣莫弗一样，拉普拉斯的发现在当时并未引起很大反响。直到十九世纪末中心极限定理的重要性才被世人所知。1901年，俄国数学家里雅普诺夫用更普通的随机变量定义中心极限定理并在数学上进行了精确的证明。如今，中心极限定理被认为是（非正式地）概率论中的首席定理。【维基百科】

如果独立同分布随机变量 $X_1, X_2, \cdots, X_n$，其均值 $\mu$ 和方差 $\sigma^2$ 存在，则其均值 $\bar{X} = \frac{1}{n}\sum X_i$ 依分布收敛于正态分布。
$$
{\sqrt {n}}({\bar {X}}_{n}-\mu )\ \xrightarrow {d} \ {\mathcal {N}}(0,\sigma ^{2})
$$

## 一元高斯分布

高斯分布也称为正态分布，其概率密度函数为
$$
p(x) = \frac{1}{\sqrt{2\pi \sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$

上式中包含两个参数 $\mu, \sigma$，参数 $\mu$ 表示均值，参数 $\sigma$ 表示标准差。均值对应正态分布的中间位置，标准差衡量了数据围绕均值分散的程度。

如果知道了高斯分布的两个参数 $\mu, \sigma$，那么任意给定一个样本数据 $x$，$p(x)$就可以计算出来。反过来，如果给定一组样本数据，但不知道其分布参数，如何估计其参数？

例如，如果想要统计学校学生的身高情况，根据中心极限定理，可以采样一部分学生的数据样本进行分析即可。

````python

import matplotlib.colors as mcolors
import itertools
color_iter = itertools.cycle(mcolors.TABLEAU_COLORS.keys())

x = np.linspace(120, 200, 200)

mean, sigma = 170, 6
y = np.random.normal(mean, sigma, size=1000)

fig, ax = plt.subplots(figsize=(16,10))

N, bins, patches = ax.hist(y, bins=12, density=True)

for patch, color in zip(patches, color_iter):
   patch.set_facecolor(color)

p = stats.norm.pdf(x, mean, sigma)
plt.plot(x, p, color='b', linewidth=3, label='u=170')
plt.plot(x, stats.norm.pdf(x, 160, 6), color='r', linewidth=3, label='u=160')
plt.legend()
````

如上图所示，那个参数更符合数据的分布？应该选择一个分布，使得样本数据出现的概率最大。这是最大似然估计。

给定独立同分布样本 $x^{(1)}, x^{(2)}, \cdots, x^{(m)}$，其似然函数
$$
L(\mu, \sigma) = \prod_{i=1}^m p(x^{(i)} \mid \mu, \sigma) = \prod_{i=1}^m \frac{1}{\sqrt{2\pi \sigma^2}} e^{-\frac{(x^{(i)}-\mu)^2}{2\sigma^2}}
$$

对数似然函数
$$
l(\mu, \sigma) = \sum_{i=1}^m \log (\frac{1}{\sqrt{2\pi \sigma^2}} e^{-\frac{(x^{(i)}-\mu)^2}{2\sigma^2}})
= -\frac{m}{2} \log 2 \pi \sigma^2 - \frac{1}{2\sigma^2} \sum_{i=1}^m(x^{(i)}-\mu)^2
$$

对数似然函数求偏导
$$
\frac{\partial l(\mu, \sigma)}{\partial \mu} = - \frac{1}{\sigma^2} \sum_{i=1}^m(x^{(i)}-\mu) = 0
$$
$$
\frac{\partial l(\mu, \sigma)}{\partial \sigma ^2} = -\frac{1}{2} \cdot \frac{m}{\sigma ^2}- \frac{1}{2} \cdot \frac{1}{\sigma^4} \sum_{i=1}^m(x^{(i)}-\mu)^2 = 0
$$

容易得到
$$
\mu = \frac{\sum_{i=1}^m x^{(i)}}{m}
$$
$$
\sigma^2 = \frac{\sum_{i=1}^m(x^{(i)}-\mu)^2}{m}
$$

根据公式，容易计算出，

````python
mu = np.sum(y) / len(y)
sigma2 = np.sum((y - mu)**2)/len(y)
sigma = np.sqrt(sigma2)
print(mu, sigma)
````

如果要进一步分别统计男生和女生的身高情况，那么可以分开两类，估算两次就可以了。

但是如果给定的数据丢失了性别信息，又如何处理？

````python
x = np.linspace(120, 200, 200)

mean_m, sigma_m = 170, 6
mean_w, sigma_w = 150, 6

X_m = np.random.normal(mean_m, sigma_m, size=2000)
X_w = np.random.normal(mean_w, sigma_w, size=2000)
fig, ax = plt.subplots(figsize=(16,10))
kwargs = dict(histtype='stepfilled', alpha=0.3, density=True, bins=40)
ax.hist(X_m, **kwargs)
ax.hist(X_w, **kwargs)
p_m = stats.norm.pdf(x, mean_m, sigma_m)
p_w = stats.norm.pdf(x, mean_w, sigma_w)
plt.plot(x, p_m, color='b', linewidth=3, label='BOY')
plt.plot(x, p_w, color='r', linewidth=3, label='GIRL')
plt.vlines(x=160, ymin=0.0, ymax=0.07, color='purple', linestyles='--')
plt.legend()
````

令 $z = 0$ 表示男生，$z=1$ 表示女生，那么，根据全概率公式，给定一个样本 $x^{(i)}$ 的概率
$$
p(x^{(i)}) = p(z=0)p(x^{(i)} \mid z = 0) +  p(z=1)p(x^{(i)} \mid z = 1) 
$$

其中 $p(z=k)$称为 $z=k(k=0,1)$ 的先验概率；$p(x^{(i)} \mid z = 0) = p(x^{(i)} \mid \mu_0, \sigma_0) $，$p(x^{(i)} \mid z = 1) = p(x^{(i)} \mid \mu_1, \sigma_1) $。

根据贝叶斯定理，$z=k$ 的后验概率
$$
p(z=0 \mid x^{(i)}) = \frac{p(z=0)p(x^{(i)} \mid z=0)}{p(x^{(i)})} = \frac{p(z=0)p(x^{(i)} \mid z=0)}{p(z=0)p(x^{(i)} \mid z = 0) +  p(z=1)p(x^{(i)} \mid z = 1) }
$$
$$
p(z=1 \mid x^{(i)}) = \frac{p(z=1)p(x^{(i)} \mid z=1)}{p(x^{(i)})} = \frac{p(z=0)p(x^{(i)} \mid z=0)}{p(z=0)p(x^{(i)} \mid z = 0) +  p(z=1)p(x^{(i)} \mid z = 1) }
$$

根据$z=k$ 的后验概率可以判断 $x^{(i)}$ 属于男生还是女生，所有数据都分好类之后，那么对于不同的类采用最大似然估计，估算出各自分布的参数以及 $z=k$ 的先验概率。然后再一次计算 $z=k$ 的后验概率......不断重复。这种方法称为期望最大值(`EM`)迭代法。

### 多元高斯分布

若 $d$ 维空间随机变量 $\boldsymbol{x} = (x_1, x_2, \cdots, x_d)$服从高斯分布，则其概率密度函数为
$$
p(\boldsymbol{x}) = \frac{1}{(2\pi)^{\frac{d}{2}}|\boldsymbol{\Sigma}|^{\frac{1}{2}}}e^{-\frac{1}{2}(\boldsymbol{x} - \boldsymbol{\mu})^{\mathrm{T}}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x} - \boldsymbol{\mu})}
$$

其中 $\boldsymbol{u} = (\mu_1, \mu_2, \cdots, \mu_d)$ 是 $d$ 维均值向量，$\boldsymbol{\Sigma}_{i,j} = E[(x_i - \mu_i)( x_j - \mu_j)] = Cov[x_i, x_j]$ 是 $d \times d$的协方差矩阵，并且 $|\boldsymbol{\Sigma}|$表示行列式。

高斯分布完全由 $\boldsymbol{u}$ 和 $\boldsymbol{\Sigma}$ 两个参数决定，所以可以记作 $p(\boldsymbol{x} \mid \boldsymbol{u}, \boldsymbol{\Sigma})$。

## 高斯混合模型

高斯混合模型（Gaussian Mixed Model，GMM）也是一种常见的聚类算法。高斯混合模型假设每个簇的数据都是符合高斯分布（又叫正态分布）的，当前数据呈现的分布就是各个簇的高斯分布叠加在一起的结果。



````python
from matplotlib.patches import Ellipse
import matplotlib.transforms as transforms

def confidence_ellipse(x, y, ax, n_std=3.0, facecolor='none', **kwargs):
    cov = np.cov(x, y)
    pearson = cov[0, 1]/np.sqrt(cov[0, 0] * cov[1, 1])
    # Using a special case to obtain the eigenvalues of this
    # two-dimensional dataset.
    ell_radius_x = np.sqrt(1 + pearson)
    ell_radius_y = np.sqrt(1 - pearson)
    ellipse = Ellipse((0, 0), width=ell_radius_x * 2, height=ell_radius_y * 2,
                      facecolor=facecolor, **kwargs)

    # Calculating the standard deviation of x from
    # the squareroot of the variance and multiplying
    # with the given number of standard deviations.
    scale_x = np.sqrt(cov[0, 0]) * n_std
    mean_x = np.mean(x)

    # calculating the standard deviation of y ...
    scale_y = np.sqrt(cov[1, 1]) * n_std
    mean_y = np.mean(y)

    transf = transforms.Affine2D() \
        .rotate_deg(45) \
        .scale(scale_x, scale_y) \
        .translate(mean_x, mean_y)

    ellipse.set_transform(transf + ax.transData)
    return ax.add_patch(ellipse)
````

````python
from sklearn.datasets import make_blobs
centers = [[3, 3], [-3, -3], [3, -3]]
X, y = make_blobs(n_samples=1000, centers=centers, n_features=2, random_state=2022)
fig, ax = plt.subplots(figsize=(12,8))
sns.scatterplot(x=X[:,0], y=X[:,1], hue=y)
X0 = X[y==0, :]
X1 = X[y==1, :]
X2 = X[y==2, :]
confidence_ellipse(X0[:,0], X0[:,1], ax, edgecolor='blue')
confidence_ellipse(X1[:,0], X1[:,1], ax, edgecolor='red')
confidence_ellipse(X2[:,0], X2[:,1], ax, edgecolor='green')
plt.axis('scaled')
````

高斯混合分布
$$
p_M(\boldsymbol{x}) = \sum_{k=0}^{K} \alpha_i \cdot p(\boldsymbol{x} \mid \boldsymbol{u}_k, \boldsymbol{\Sigma}_k)
$$
表示由 $K$ 个混合成分组成，每个混合成分对应一个高斯分布，$\boldsymbol{u}_k, \boldsymbol{\Sigma}_k$ 是其对应的参数，$\alpha_k$ 是对应的混合系数，且满足 $\sum_{k=0}^{K} \alpha_k = 1$。

令 $z^{(j)} \in \{1, 2, \cdots, K\}$表示生成样本 $\boldsymbol{x}^{(j)}$ 的高斯混合成分，或者说是样本的类别，显然 $p(z^{(j)} = k) = \alpha_k$。那么
$$
p(z^{(j)}= k \mid \boldsymbol{x}^{(j)}) = \frac{\alpha_k \cdot p(\boldsymbol{x}^{(j)} \mid \boldsymbol{u}_k, \boldsymbol{\Sigma}_k)} {\sum_{k=0}^{K} \alpha_k \cdot p(\boldsymbol{x}^{(j)} \mid \boldsymbol{u}_k, \boldsymbol{\Sigma}_k)}
$$

$p(z^{(j)}= k \mid \boldsymbol{x}^{(j)})$ 给出了样本由第 $k$ 个高斯混合成分生成的后验概率，通常记作 $\gamma_{jk}$。如果知道了参数 $\boldsymbol{u}_i, \boldsymbol{\Sigma}_i$，那么，给定一个样本  $\boldsymbol{x}^{(j)}$，要判断其属于哪一个簇或类别，只需要找到后验概率最大的那一个成分就好了，即样本$\boldsymbol{x}^{(j)}$的类别 $\lambda^{(j)}$为
$$
\lambda ^{(j)} = \argmax_{k\in \{1, 2, \cdots, K\}} \gamma_{jk}
$$

现在的问题是如何计算高斯混合模型的参数 $\boldsymbol{u}_k, \boldsymbol{\Sigma}_k$。显然，已经知道了样本的分布，那么可以采用极大似然估计，即最大化
$$
L =  \log (\prod_{j=1}^m  p_M(\boldsymbol{x}^{(j)})) =  \sum_{j=1}^m\log (\sum_{k=0}^{K} \alpha_k \cdot p(\boldsymbol{x}^{(j)} \mid \boldsymbol{u}_k, \boldsymbol{\Sigma}_k)) 
$$

上面式子还存在一个参数 $\alpha_k$，称为隐变量，所以一般采用EM(Expectation Maximumn)算法进行迭代求解。

对 $L$ 求偏导，容易得到
$$
\boldsymbol{\mu}_k = \frac{\sum_{j=1}^m \gamma_{jk} \boldsymbol{x}^{(j)}}{\sum_{j=1}^m \gamma_{jk}}
$$

$$
\boldsymbol{\Sigma}_k = \frac{\sum_{j=1}^m \gamma_{jk} (\boldsymbol{x}^{(j)} - \boldsymbol{\mu}_k)(\boldsymbol{x}^{(j)} - \boldsymbol{\mu}_k)^{\mathrm{T}} }{\sum_{j=1}^m \gamma_{jk}}
$$

而系数 $\alpha_k$，需要满足约束 $\alpha_k \ge 0$， 并且 $\sum_{k=0}^{K} \alpha_k = 1$，容易得到拉格朗日形式
$$
L + \lambda (\sum_{k=0}^{K} \alpha_k - 1)
$$
上述对 $\alpha_k$求导，得到
$$
\sum_{j=1}^{m} \gamma_{jk} + \lambda \alpha_k = 0
$$
上式对所有 $k$ 求和可知 $\lambda = -m$, 所以
$$
\alpha_k = \frac{1}{m} \sum_{j=1}^{m} \gamma_{jk}
$$

### `GMM`算法流程

输入：样本集 $D$，聚类的簇数$K$

输出：簇划分 $C = \{C_1, C_2, \cdots, C_K \}$.

1. 初始化参数 $\alpha, \mu, \Sigma$。

2. 计算 $\gamma$。（E步）
$$
p(\boldsymbol{x}) = \frac{1}{(2\pi)^{\frac{d}{2}}|\boldsymbol{\Sigma}|^{\frac{1}{2}}}e^{-\frac{1}{2}(\boldsymbol{x} - \boldsymbol{\mu})^{\mathrm{T}}\boldsymbol{\Sigma}^{-1}(\boldsymbol{x} - \boldsymbol{\mu})}
$$
$$
\gamma_{jk} = \frac{\alpha_k \cdot p(\boldsymbol{x}^{(j)} \mid \boldsymbol{u}_k, \boldsymbol{\Sigma}_k)} {\sum_{k=0}^{K} \alpha_k \cdot p(\boldsymbol{x}^{(j)} \mid \boldsymbol{u}_k, \boldsymbol{\Sigma}_k)}
$$

3. 更新 $\alpha, \mu, \Sigma$。（M步）

$$
\alpha_k = \frac{1}{m} \sum_{j=1}^{m} \gamma_{jk}
$$

$$
\boldsymbol{\mu}_k = \frac{\sum_{j=1}^m \gamma_{jk} \boldsymbol{x}^{(j)}}{\sum_{j=1}^m \gamma_{jk}}
$$

$$
\boldsymbol{\Sigma}_k = \frac{\sum_{j=1}^m \gamma_{jk} (\boldsymbol{x}^{(j)} - \boldsymbol{\mu}_k)(\boldsymbol{x}^{(j)} - \boldsymbol{\mu}_k)^{\mathrm{T}} }{\sum_{j=1}^m \gamma_{jk}}
$$



4. 重复步骤2和步骤3。

5. 直到达到收敛或者达到某个停止阈值(如最大计算时间)。

````python
def multi_gaussian(X, mu, sigma):
    """
    input
        x:     m*d
        mu:    1*d
        sigma: d*d
    return 
        prob: m*1
    """
    def gaussian(x):
        return (1.0 / np.sqrt((2*np.pi)**sigma.shape[0] * np.linalg.det(sigma))) * np.exp(-0.5 * (x - mu) @ np.linalg.pinv(sigma) @ (x - mu).T)
    return np.apply_along_axis(gaussian, axis=1, arr=X)

def init(m, d, K):
    """
    return 
        mu:    k*d
        sigma: d*d
        alpha: k*1
    """
    mu = np.random.randn(K, d)
    sigma = np.tile(np.eye(d), (K, 1, 1))
    alpha = np.full((K, 1), 1.0 / K)
    return mu, sigma, alpha

def e_step(X, mu, sigma, alpha):
    """
    input
        X:     m*d
        mu:    K*d
        sigma: K*d*d
        alpha: K*1
    return
        gamma: m*k
    """
    prob = np.array([multi_gaussian(X, mu[i], sigma[i]) for i in range(mu.shape[0])]).T
    gamma =  prob * alpha.T
    gamma = gamma / (gamma.sum(axis=1)[:,None])

    return gamma

def m_step(X, gamma):
    """
    input
        X:     m*d
        gamma: m*K
    return 
        mu:    k*d
        sigma: d*d
        alpha: k*1
    """
    m, K = X.shape[0], gamma.shape[1]
    gamma_sum = gamma.sum(axis=0).reshape(-1, 1)
    alpha = gamma_sum / m
    mu = gamma.T @ X / gamma_sum
    sigma = np.array([((gamma[:, k, None] * (X - mu[k])).T @  (X - mu[k])) / gamma_sum[k] for k in range(K)])
    return mu, sigma, alpha

def gmm(X, K, epoches):
    """
    input
        X:       m*d
        K:       int
        epoches: int
    return
        C:       m*1
    """
    m, d = X.shape
    mu, sigma, alpha = init(m, d, K)
    for _ in range(epoches):
        gamma = e_step(X, mu, sigma, alpha)
        mu, sigma, alpha = m_step(X, gamma)

    return gamma.argmax(axis=1)
    
````

````python
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=1000, centers=4, cluster_std=0.60, random_state=0)
y_pred = gmm(X, 4, 100)

plt.figure()
ax = plt.scatter(X[:, 0], X[:, 1], c=y_true)
plt.legend(*ax.legend_elements(), loc="upper right", title="Classes")
plt.title("Ground True")

plt.figure()
ax = plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.legend(*ax.legend_elements(), loc="upper right", title="Classes")
plt.title("Prediction")
````

### 参数

- `n_components`：混合高斯模型个数，默认为 1 。

- `covariance_type`：协方差类型，包括 `{'full','tied', 'diag', 'spherical'}` 四种，默认`'full' `完全协方差矩阵。

    - `full` 指每个分量有各自不同的标准协方差矩阵，完全协方差矩阵（元素都不为零）；

    - `tied` 指所有分量有相同的标准协方差矩阵（`HMM`会用到）；

    - `diag` 指每个分量有各自不同对角协方差矩阵（非对角为零，对角不为零）；
    
    - `spherical` 指每个分量有各自不同的简单协方差矩阵，球面协方差矩阵（非对角为零，对角完全相同，球面特性）。

- `tol`：迭代停止阈值，默认为 `1e-3`。

- `reg_covar`：协方差对角非负正则化，保证协方差矩阵均为正，默认为 0 。

- `max_iter`：最大迭代次数，默认 100 。

- `n_init`：初始化次数，用于产生最佳初始参数，默认为 1 。

- `init_params`：`{'kmeans', 'random'}`，初始化参数实现方式，默认用 `kmeans` 实现，也可以选择随机产生。 

- `weights_init`：各组成模型的先验权重，如果没有给定，那么使用`init_params`参数给定的方法来进行创建，默认为`None`。

- `means_init`：初始化均值，如果没有给定，那么使用`init_params`参数给定的方法来进行创建，默认为`None`。

- `precisions_init`：初始化精确度（模型个数，特征个数），如果没有给定，那么使用`init_params`参数给定的方法来进行创建，默认为`None`，它的形状依靠协方差类型。

- `random_state` ：随机数发生器 

- `warm_start `：布尔类型，默认为`False`，当该值为`True`的时候，最后一次的学习结果作为下一次调用的初始化参数，在类似问题被多次训练的时候，可以加快收敛速度。

- `verbose`：`int`类型, 默认为0。作用：可以冗余输出，如果为1，它会打印正确的初始化结果和每一次迭代步骤。如果大于1，则打印日志概率和每一步所需的时间。

- `verbose_interval`： `int`类型, 默认为 10。作用：下一次打印之前的迭代发生次数。

### 属性

- `weights_` ： `array-like, shape (n_components,)`，每个混合元素权重。

- `means_` ： `array-like, shape (n_components, n_features)`，每个混合元素均值。

- `covariances_` ： `array-like`，每个混合元素的协方差，它的形状依靠协方差类型：

    协方差类型 |	协方差形状
    |:---|:--|
    `'spherical'` |	`(n_components,)`
    `'tied'`	| `(n_features, n_features)`
    `'diag'`	 | `(n_components, n_features)`
    `'full'` |	`(n_components, n_features, n_features)`


- `precisions_` ： `array-like`，每个混合元素精度矩阵，精度矩阵是协方差矩阵的逆。协方差矩阵是对称正定的，因此高斯混合可以用精度矩阵等价地参数化。存储精度矩阵而不是协方差矩阵使得在测试时间计算新样本的对数似然更有效。它的形状依靠协方差类型，同协方差一样。

- `converged_ `： `bool`类型，在`fit()`中达到收敛时为`True`，否则为`False`。

- `lower_bound_` ： `float`类型，EM最佳拟合的对数似然(与模型相关的训练数据)的下限值。

### 方法

- `aic(self, X)	`

    在当前模型上输入`X`的Akaike信息准则

- `bic(self, X)	`

    在当前模型上输入`X`的贝叶斯信息准则。

- `fit(self, X[, y])`

    用EM算法估计模型参数。

- `fit_predict(self, X[, y])`

    用`X`估计模型参数，并预测`X`的标签。

- `get_params(self[, deep])`	

    获取这个算法的参数

- `predict(self, X)`	

    使用训练模型预测`X`中数据样本的标签。

- `predict_proba(self, X)`

    在给定数据的情况下，预测每个分量的后验概率。

- `sample(self[, n_samples])	`

    从拟合的高斯分布生成随机样本。

- `score(self, X[, y])	`

    计算给定数据`X`的每个样本平均`log`似然函数

- `score_samples(self, X)	`

    计算每个样本的加权`log`概率。

- `set_params(self, **params)`	

    为算法设置参数
